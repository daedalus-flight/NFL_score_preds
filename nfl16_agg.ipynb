#### - note that the rolling average shifted 1 to reflect information that week

```python
def make_team_df(team):
    team = pd.DataFrame(nfl16[nfl16['team_x'] == team])
    team['pf_5mva'] = team.iloc[:,4].rolling(window=5).mean().shift()
    team['pa_5mva'] = team.iloc[:,5].rolling(window=5).mean().shift()
    team['margin_5mva'] = team.iloc[:,6].rolling(window=5).mean().shift()
    team['pyards_5mva'] = team.iloc[:,10].rolling(window=5).mean().shift()
    team['ryards_5mva'] = team.iloc[:,11].rolling(window=5).mean().shift()
    team.dropna(inplace=True)
    return team
```
