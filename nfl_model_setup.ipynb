```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None) 
```

```python
nfl_ma = pd.read_csv('./data/nfl_fin.csv')
```

```python
nfl_ma.tail()
```

```python
def make_team_df(team, mva=7):
    team = pd.DataFrame(nfl_ma[nfl_ma['target_team'] == 'NYJ'])
    team[f'pf_{mva}mva'] = team.iloc[:,5].rolling(window=mva).mean().shift()
    team[f'pa_{mva}mva'] = team.iloc[:,6].rolling(window=mva).mean().shift()
    team[f'margin_{mva}mva'] = team.iloc[:,8].rolling(window=mva).mean().shift()
    team[f'pyards_{mva}mva'] = team.iloc[:,9].rolling(window=mva).mean().shift()
    team[f'ryards_{mva}mva'] = team.iloc[:,10].rolling(window=mva).mean().shift()
    team[f'opp_pyards_{mva}mva'] = team.iloc[:,11].rolling(window=mva).mean().shift()
    team[f'opp_ryards_{mva}mva'] = team.iloc[:,12].rolling(window=mva).mean().shift()
    team.dropna(inplace=True)
    return team
```

```python
nfl7mva = pd.concat([car,
    den,
    ari,
    atl,
    bal,
    buf,
    chi,
    cin,
    cle,
    dal,
    det,
    gb,
    hou,
    ind,
    jax,
    kc,
    mia,
    min,
    ne,
    no,
    nyg,
    nyj,
    oak,
    phi,
    sdg,
    sea,
    tb,
    ten,
    lar,
    pit,
    sf,
    was,], axis= 0, join='inner')
```
